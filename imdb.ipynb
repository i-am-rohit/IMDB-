```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from random import randint
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
}
url = "https://www.imdb.com/list/ls063676189/?sort=list_order,asc&st_dt=&mode=detail&page=1&ref_=ttls_vm_dtl"
```

```python
for page_number in range(0, 49):  # Loop for 5 pages
  page_url = url + str(page_number)
  print(page_url)
```

```python
response = requests.get(page_url, headers=headers)
response 
```

```python
try:
    response = requests.get(page_url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text,'html.parser')
    # print(soup)
    movies = soup.find_all(class_="lister-item mode-detail")
    print(len(movies))
except Exception as e:
    print(e)
```

```python
for movie in movies:
    name = movie.find(class_ = "lister-item-header")
    print(name)
    break
```

```python
Ranking = []
Name = []
Director = []
actor = []
Year = []
Imdb_rating = []
metascore = []
Runtime = []
Genre = []
Voting = []
Gross_collection = []
description = []
Image = []


for movie in movies:

    rank = movie.find('span', class_ = 'lister-item-index unbold text-primary').text.split(".")[0]
    Ranking.append(rank)
    
    name = movie.find(class_ = "lister-item-header").a.text
    Name.append(name)

    value0 = movie.find_all(class_ = "text-muted text-small")

    Direct  = value0[1].a.text
    Director.append(Direct)

    

    
    year = movie.find(class_ = "lister-item-year text-muted unbold").string.strip('()')
    Year.append(year)
    
    rating = movie.find(class_ = "ipl-rating-star__rating").string      
    Imdb_rating.append(rating)
    
    m_score = movie.find('span', class_ = 'metascore').text.strip(' ')
    metascore.append(m_score)

    Run = movie.find(class_="runtime").text
    Runtime.append(Run)

    genre = movie.find(class_="genre").string.strip(' ').replace('\n',' ')
    Genre.append(genre)   

    # desc = movie.find(class_="text-muted text-small").text  if movie.find(class_="text-muted text-small") else("Not Present")

    value1 = movie.find_all('span', attrs={'name' : 'nv'})
    
    vote = value1[0].text
    Voting.append(vote)

    Gross = value1[1].text
    Gross_collection.append(Gross)


    desc = movie.find("p", class_ = '')
    desc = desc.text.replace('\n', '').split('|')
    # desc = [x.strip() for x in cast]
    description.append(desc)
   

    images = movie.find('img', class_="loadlate")
    image_url = images.get('loadlate')
    Image.append(image_url)



    cast = value0[1].text.split('|')
    cast = cast[1].split(':')
    actor.append(cast[1].replace('\n',' '))
    
    print(rank,    ","    , image_url ,  ","    , name,  ","     ,   Director , ","  , actor  ,     ","       ,   year ,  ","   , rating ,  ","  ,  m_score ,   ","  ,  Run,   ","  ,  genre ,  ","  , vote ,  ","  , Gross ,  ","  , desc )
    

```

```python
movie_list = pd.DataFrame({"Ranking" : Ranking, "Movie" : Name,"Description" : description ,"Image" :image_url, 'Director': Director, "Cast" : actor, "Year": year,"IMDB_rating": Imdb_rating, "Meta_score":metascore, "Runtime": Runtime, "Genre": genre,
                          "Gross_collection": Gross_collection, "Voting" : Voting})
```
